```python
'''Restaurant Billing Logic

Ask user:

amount

loyalty member (yes/no)

If amount > 1000 and loyalty = yes ‚Üí 20% discount Else if amount > 1000 ‚Üí 10% discount Else ‚Üí no discount'''

Amount=int(input("enter the amount"))
loyalty=input("enter the loyalty in yes or no")
if Amount>1000 and loyalty=="yes":
  print("20% discount")
  print(Amount*0.8)
elif Amount >100 and loyalty=="no":
  print("10% discount")
  print(Amount*0.9)
else:
  print(" no discount")
  print(Amount)
```

```python
'''Customer Churn Prediction (symbolic logic)

Input:

complaints ‚â• 3

monthly_usage < 10 GB

payment_delay > 2 times

If any 2 conditions are true ‚Üí print ‚ÄúLikely to Churn‚Äù Else ‚Üí ‚ÄúSafe Customer‚Äù. '''

complaints=int(input("enter the  no of complaints"))
monthly_usage=int(input("enter the monthly usage"))
payment_delay=int(input("enter the no of payment delay"))
if (complaints>=3 and monthly_usage<10) or (complaints>=3 and payment_delay>2) or (monthly_usage<10 and payment_delay>2):
  print("likely to churn")
else:
  print("safe")

```
